```python
from tqdm import tqdm
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import xarray as xr
```

```python
%matplotlib inline
```

```python
ds = xr.open_dataset('/home/cz2397/data/rclv-eddies/traj_0000-0090.nc')
ds
```

```python
lon = ds.x
lat = ds.y
```

```python
lon_eddy = np.zeros((10, 90))
lat_eddy = np.zeros((10, 90))
```

```python
for i in tqdm(range(10)):
    for j in range(90):
        lon_eddy[i][j] = lon[j][i]
        lat_eddy[i][j] = lat[j][i]
```

```python
# npart 1
# lifetime 90 days
sns.set()
plt.plot(lon_eddy[0], lat_eddy[0])
```

```python
sns.set()
plt.plot(lon_eddy[1], lat_eddy[1])
```

```python
sns.set()
plt.plot(lon_eddy[2], lat_eddy[2])
```
